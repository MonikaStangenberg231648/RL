```python
from stable_baselines3 import PPO
from ot2_gym_wrapper import OT2Env  # Import rodowiska OT2Env

# Tworzenie rodowiska OT2
env = OT2Env(render=False)  # render=False, aby przyspieszy trening

# Inicjalizacja modelu PPO z hiperparametrami
model = PPO(
    policy='MlpPolicy', 
    env=env, 
    verbose=1,             # Wywietlanie logw
    learning_rate=0.0003,  # Szybko uczenia
    n_steps=2048,          # Liczba krokw przed aktualizacj polityki
    batch_size=64,         # Rozmiar batcha dla gradientu
    n_epochs=10            # Liczba epok na batch
)

# Trening modelu przez okrelon liczb krokw
total_timesteps = 10000  # Liczba krokw do treningu (dostosuj, jeli potrzebujesz szybszych testw)
model.learn(total_timesteps=total_timesteps, progress_bar=True)

# Zapisywanie wytrenowanego modelu
model.save("ppo_ot2_model")

# adowanie modelu (opcjonalnie, do testowania lub dalszego treningu)
loaded_model = PPO.load("ppo_ot2_model")

# Testowanie modelu po treningu
obs = env.reset()  # Resetowanie rodowiska
done = False

while not done:
    # Przewidywanie akcji na podstawie aktualnej obserwacji
    action, _ = loaded_model.predict(obs)
    # Wykonanie akcji w rodowisku
    obs, reward, done, info = env.step(action)

```
