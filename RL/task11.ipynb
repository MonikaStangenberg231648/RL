```python
from stable_baselines3 import PPO
from ot2_gym_wrapper import OT2Env  # Import środowiska OT2Env

# Tworzenie środowiska OT2
env = OT2Env(render=False)  # render=False, aby przyspieszyć trening

# Inicjalizacja modelu PPO z hiperparametrami
model = PPO(
    policy='MlpPolicy', 
    env=env, 
    verbose=1,             # Wyświetlanie logów
    learning_rate=0.0003,  # Szybkość uczenia
    n_steps=2048,          # Liczba kroków przed aktualizacją polityki
    batch_size=64,         # Rozmiar batcha dla gradientu
    n_epochs=10            # Liczba epok na batch
)

# Trening modelu przez określoną liczbę kroków
total_timesteps = 10000  # Liczba kroków do treningu (dostosuj, jeśli potrzebujesz szybszych testów)
model.learn(total_timesteps=total_timesteps, progress_bar=True)

# Zapisywanie wytrenowanego modelu
model.save("ppo_ot2_model")

# Ładowanie modelu (opcjonalnie, do testowania lub dalszego treningu)
loaded_model = PPO.load("ppo_ot2_model")

# Testowanie modelu po treningu
obs = env.reset()  # Resetowanie środowiska
done = False

while not done:
    # Przewidywanie akcji na podstawie aktualnej obserwacji
    action, _ = loaded_model.predict(obs)
    # Wykonanie akcji w środowisku
    obs, reward, done, info = env.step(action)

```
